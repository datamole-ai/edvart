```python
import edvart

import plotly.offline as py
py.init_notebook_mode()
```

```python
dataset_ts = edvart.example_datasets.dataset_global_temp()
```

```python
report_ts = edvart.DefaultTimeseriesReport(
    dataset_ts,
    # Monthly data -> analyze yearly seasonality
    sampling_rate=12,
    stft_window_size=24,
)
```

```python
report_ts.show()
```

```python
report_ts.export_html("time-series-report-exported.html")
```

```python
report_ts.export_notebook("time-series-report-exported.ipynb")
```
